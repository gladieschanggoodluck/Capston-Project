```python
df1 = df[['numPages', 'height', 'width', 'dim', 'structure', 'content', 'association', 'numChar', 'researchPublication']]
```

```python
df2 = df1.values
```

```python
# Seperating out the features
x = df2[:, 1:9].astype(float)
# Seperating out te target
y = df2[:, -1]
# Standardizing the features
x = StandardScaler().fit_transform(x)
```

# PCA Projection to 2D

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

```python
finalDf = pd.concat([principalDf, df1[['researchPublication']]], axis = 1)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
researchPublication = [1, 0]
colors = ['r', 'g']
for researchPublication, color in zip(researchPublication,colors):
    indicesToKeep = finalDf['researchPublication'] == researchPublication
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend('researchPublication')
ax.grid()
```
